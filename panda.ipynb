# handling null values

```python
import pandas as pd
```

```python
ufo=pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
ufo.isnull().sum()
```

```python
# subset is to select few columns for excluding null values - " how" determines if the funtion needs to check null values in all or any few columns
ufo.dropna(subset=['country','continent'],how='any').shape
```

```python
ufo['country'].value_counts
```
