#Determining college campus diversity

```python
college=pd.read_csv('college.csv',index_col='INSTNM')
```

```python
college.columns
```

```python
college.describe()
```

```python
college.columns
```

```python
first=college.filter(like='UGDS_')
first.columns
```

```python
first.head(5)
```

```python
first.isnull().sum(axis=1).sort_values(ascending=False).head()
first
```

```python
first.dropna(how='all')
first
```

```python
first.isnull().sum(axis=0)
first
```

```python
diversity=first.ge(.15).sum(axis='columns').sort_values(ascending=False)
diversity
```
