```python
import re
import pandas as pd
```

```python
r_dict={
    'Client':re.compile(r'.CssrCardProvider userDetails: \[{"client":\[("[\w\d]{10}","[\w\d]{10}","[\w\d]{10}","[\w\d]{10}","[\w\d]{10}","[\w\d]{10}","[\w\d]{10}")\]'),
    'Persona':re.compile(r'.CssrCardProvider userDetails: \[{"client":\["[\w\d]{10}","[\w\d]{10}","[\w\d]{10}","[\w\d]{10}","[\w\d]{10}","[\w\d]{10}","[\w\d]{10}"\],"role":\[\],"persona":\[("[\w]{3}")\]')
}
```

```python
def p_line(line):
    for key,rx in r_dict.items():
        match=rx.search(line)
        if match:
            return key,match
        return None, None
```

```python
def parse_file(filepath):
    data=[]
    with open(filepath,'r') as file_object:
        line=file_object.readline()
        while line:
            key,match = p_line(line)
            if key == 'client':
                client=match.group('Client')
            if key == 'persona':
                persona=match.group('Persona')
            row={'Client':client,'Persona':persona}
            data.append(row)
            line=file_object.readline()
        line=file_object.readline()
        data=pd.DataFrame(data)
    return data
        
```

```python
filepath='ITCONSOLELOG.txt'
data=parse_file(filepath)
print(data)
```
